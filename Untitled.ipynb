We have confirmed the outliers in the rainfall as above. This helps us to judge if they are extreme weather conditions or errors in data recording.

After carefully studying the data, we concluded they were extreme conditions and hence we will be applying log transformation on them to help minimise impact of large outliers and stabilize variance. This will make our data more suitable for modelling.

```python
if (agri_df['Rainfall'] <= 0).any():
    agri_df['Rainfall'] = agri_df['Rainfall'] + 1
# Apply log transformation
agri_df['Log_Rainfall'] = np.log(agri_df['Rainfall'])

# Visualize the original and transformed data
plt.figure(figsize=(7, 4))
# Original Rainfall
plt.subplot(1, 2, 1)
plt.hist(agri_df['Rainfall'], bins=30, color='skyblue', edgecolor='black')
plt.title('Original Rainfall Data')
plt.xlabel('Rainfall')
plt.ylabel('Frequency')
# Log-transformed Rainfall
plt.subplot(1, 2, 2)
plt.hist(agri_df['Log_Rainfall'], bins=30, color='salmon', edgecolor='black')
plt.title('Log-Transformed Rainfall Data')
plt.xlabel('Log(Rainfall)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

<h3>Model Development</h3>

To avoid Multicollinearity we will  be using the Month Average and and Year as our Features variable so as to predict the Log_Rainfall and temprature.

<h5> Split Data</h5>

```python
# feature and targets variables.
X = agri_df[['Year', 'Month Average']]
y_rainfall = agri_df['Log_Rainfall']
y_temperature = agri_df['Temperature']

# Train-test split
X_train, X_test, y_train_rainfall, y_test_rainfall = train_test_split(X, y_rainfall, test_size=0.2, random_state=42)
X_train_temp, X_test_temp, y_train_temp, y_test_temp = train_test_split(X, y_temperature, test_size=0.2, random_state=42)
```

<h5>LinearRegression model for base analysis</h5>

```python
# Regression for Rainfall
model_rainfall = LinearRegression()
model_rainfall.fit(X_train, y_train_rainfall)
# Rigression for Temperature
model_temperature = LinearRegression()
model_temperature.fit(X_train_temp, y_train_temp)

# Prediction
y_pred_rainfall = model_rainfall.predict(X_test)
y_pred_temperature = model_temperature.predict(X_test_temp)

# Evaluating MSE and R2 Score
rainfall_mse = mean_squared_error(y_test_rainfall, y_pred_rainfall)
temperature_mse = mean_squared_error(y_test_temp, y_pred_temperature)
rainfall_r2 = r2_score(y_test_rainfall, y_pred_rainfall)
temperature_r2 = r2_score(y_test_temp, y_pred_temperature)

print(f"Rainfall MSE: {rainfall_mse}, R^2: {rainfall_r2}")
print(f"Temperature MSE: {temperature_mse}, R^2: {temperature_r2}")
```

According to our linear regression model, it is approximately 4.6% accurate on rainfall prediction and 25.2% accurate on temperature. this acts as our base model but we will not be using it since we are dealing with non-linear data.

<h4>RandomForestRegressor.</h4>

<h4> For Rainfall</h4>

```python
#  Model
rf_rainfall = RandomForestRegressor(random_state=42)

# parameters
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [1.0],
    'bootstrap': [True, False]
}
# GridSearch with cross-validation
grid_search = GridSearchCV(estimator=rf_rainfall, param_grid=param_grid, 
                           scoring='neg_mean_squared_error', 
                           cv=5, n_jobs=-1, verbose=2)
# fitting
grid_search.fit(X_train, y_train_rainfall)

# Best parameters
best_params = grid_search.best_params_
print("Best parameters for Rainfall Random Forest:", best_params)
# Cross-validated MSE score
cv_mse_scores = -grid_search.cv_results_['mean_test_score']
mean_cv_mse = np.mean(cv_mse_scores)
print("Mean Cross-validated MSE:", mean_cv_mse)

# Prediction on test data
best_rf_rainfall = grid_search.best_estimator_
y_pred_rf_rainfall = best_rf_rainfall.predict(X_test)
# Evaluation metrics
test_mse = mean_squared_error(y_test_rainfall, y_pred_rf_rainfall)
test_r2 = r2_score(y_test_rainfall, y_pred_rf_rainfall)

print("Test Set Rainfall MSE:", test_mse)
print("R²:", test_r2)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(agri_df[['Year', 'Month Average', 'Log_Rainfall', 'Temperature']])

scaled_df = pd.DataFrame(scaled_data, columns=['Year', 'Month Average', 'Log_Rainfall', 'Temperature'])

# Function 4 sequences
def create_sequences(data, target_col, time_steps):
    sequences = []
    targets = []
    for i in range(len(data) - time_steps):
        seq = data[i:i + time_steps]
        target = data[i + time_steps][target_col]
        sequences.append(seq)
        targets.append(target)
    return np.array(sequences), np.array(targets)

# Parameters
time_steps = 12 
target_col_index = 2  

# sequences 4 rainfall
X, y = create_sequences(scaled_df.values, target_col_index, time_steps)

# train and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(1))  # Output layer for regression

model.compile(optimizer='adam', loss='mean_squared_error')

# Fit 
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
# predictions
y_pred = model.predict(X_test)

y_pred_inverse = scaler.inverse_transform(np.hstack((X_test[:, -1, :2], y_pred)))[:, 2]
y_test_inverse = scaler.inverse_transform(np.hstack((X_test[:, -1, :2], y_test.reshape(-1, 1))))[:, 2]

mse = mean_squared_error(y_test_inverse, y_pred_inverse)
r2 = r2_score(y_test_inverse, y_pred_inverse)

print(f'Test Set MSE: {mse}, R²: {r2}')

import time
start_time = time.time()
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
end_time = time.time()
print(f"Training time: {end_time - start_time} seconds")
```
